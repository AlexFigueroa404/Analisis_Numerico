```python
from sympy import *
from math import *
import cmath
from tabulate import tabulate
```

```python
def tolerancia(cifras):
    return 0.5*pow(10,2-cifras)
```

```python
def errorAproximacion(valorActual,valorAnterior):
    try:
        return abs((valorActual - valorAnterior)/(valorActual))*100
    except:
        return -1
```

# Series

```python
def taylor_Exponencial(x,cifras):
    
    tabla = []
    tabla.append(["Iteracion","Valor Aproximado","Ea"])
    valor = float(x)
    Es = tolerancia(cifras)
    Ea = 1.0
    iteracion = 0
    respuesta = 0.0
    
    while Ea > Es:
        
        respuesta+= (x**iteracion)/(factorial(iteracion))        
        if iteracion > 0:
            Ea = errorAproximacion(respuesta,respuestaAnterior)
            tabla.append([iteracion+1,respuesta,Ea])
        else:
            tabla.append([iteracion+1,respuesta,"NaN"])
        iteracion+=1
        respuestaAnterior = respuesta
        
    return tabla
```

```python
def taylor_Cos(x,cifras):
    valor = radians(x)
    tabla = []
    tabla.append(["Iteracion","Valor Aproximado","Ea"])
    Es = tolerancia(cifras)
    Ea = 1
    iteracion = 0
    respuesta = 0
    while Ea > Es:
        respuesta += (pow(-1,iteracion)/(factorial(2*iteracion)))*pow(x,2*iteracion)
        
        if iteracion > 0 :
            Ea = errorAproximacion(respuesta,valorAnterior)
            if Ea != -1:
                tabla.append([iteracion+1,respuesta,Ea])
        else:
            tabla.append([iteracion+1,respuesta,"NaN"])
        valorAnterior = respuesta
        iteracion+=1
    
    return tabla
```

```python
def taylor_Seno(x,cifras):
    valor = radians(x)
    tabla = []
    tabla.append(["Iteracion","Valor Aproximado","Ea"])
    Es = tolerancia(cifras)
    Ea = 1
    iteracion = 0
    respuesta = 0
    while Ea > Es:
        respuesta += ((pow(-1,iteracion))/(factorial(2*iteracion+1)))*pow(x,2*iteracion+1)
        
        if iteracion > 0:
            Ea = errorAproximacion(respuesta,valorAnterior)
            if Ea != -1:
                tabla.append([iteracion+1,respuesta,Ea])
        else:
            tabla.append([iteracion+1,respuesta,"NaN"])
        
        valorAnterior = respuesta
        iteracion+=1
    
    return tabla
```

```python
def taylor_SenoHiperbolico(x,cifras):
    tabla = []
    tabla.append(["Iteracion","Valor Aproximado","Ea"])
    valor = radians(x)
    Es = tolerancia(cifras)
    Ea = 1
    iteracion = 0
    respuesta = 0
    while Ea > Es:
        
        respuesta += ((1)/(factorial(2*iteracion+1)))*pow(x,2*iteracion+1)
        
        if iteracion > 0:
            
            Ea = errorAproximacion(respuesta,valorAnterior)
            if Ea != -1:
                tabla.append([iteracion+1,respuesta,Ea])
        else:
            tabla.append([iteracion+1,respuesta,"NaN"])
        
        iteracion+=1
        valorAnterior=respuesta
    
    return tabla
```

```python
def taylor_CosenoHiperbolico(x, cifras):
    tabla = []
    tabla.append(["Iteracion","Valor Aproximado","Ea"])
    valor = radians(x)
    Es = tolerancia(cifras)
    Ea = 1
    iteracion = 0
    respuesta = 0
    
    while Ea > Es:
        respuesta+= ((1)/(factorial(2*iteracion)))*pow(x,2*iteracion)
        if iteracion > 0:
            Ea = errorAproximacion(respuesta,valorAnterior)
            if Ea != -1:
                tabla.append([iteracion,respuesta,Ea])
        else:
            tabla.append([iteracion,respuesta,"NaN"])
        iteracion+=1
        valorAnterior = respuesta
    return tabla
```

```python
def taylor_Ln(x,cifras):
    tabla = []
    tabla.append(["Iteracion","Valor Aproximado","Ea"])
    valor = radians(x)
    Es = tolerancia(cifras)
    Ea = 1
    iteracion = 1
    respuesta = 0.0
    
    while Ea > Es:
        respuesta+= ((pow(-1,iteracion-1))/(iteracion))*pow(x,iteracion)
        if iteracion > 1:
            Ea = errorAproximacion(respuesta,valorAnterior)
            if Ea != -1:
                tabla.append([iteracion,respuesta,Ea])
            else:
                print("error")
        else:
            tabla.append([iteracion,respuesta,"NaN"])
        
        iteracion+=1
        valorAnterior = respuesta
    return tabla
```
