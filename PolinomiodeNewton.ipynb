```python
from sympy import*
from tabulate import tabulate

x = symbols('x')


def polinomonewton(fx, x0, x1, x2, x3, x4, x5, y0, y1, y2, y3, y4, y5, grado, interpo):
    b0, b1, b2, b3, b4, b5, resultado = 0, 0, 0, 0, 0, 0, 0
    resultado = []
    resultado.append([x0, x1, x2, x3, x4, x5])
    resultado.append([y0, y1, y2, y3, y4, y5])
    if fx == 0:
        try:
            b0 = y0
            b1 = ((y1 - y0) / (x1 - x0))
            fx2x1 = (y2 - y1) / (x2 - x1)
            b2 = ((fx2x1 - b1) / (x2 - x0))
            fx3x2 = (y3 - y2) / (x3 - x2)
            fx3x2x1 = (fx3x2 - fx2x1) / (x3 - x1)
            b3 = ((fx3x2x1 - b2) / (x3 - x0))
            fx4x3 = (y4 - y3) / (x4 - x3)
            fx4x3x2 = (fx4x3 - fx3x2) / (x4 - x2)
            fx4x3x2x1 = (fx4x3x2 - fx3x2x1) / (x4 - x1)
            b4 = ((fx4x3x2x1 - b3) / (x4 - x0))
            fx5x4 = (y5 - y4) / x5 - x4
            fx5x4x3 = (fx5x4 - fx4x3) / (x5 - x3)
            fx5x4x3x2 = (fx5x4x3 - fx4x3x2) / (x5 - x2)
            fx5x4x3x2x1 = (fx5x4x3x2 - fx4x3x2x1) / (x5 - x1)
            b5 = (fx5x4x3x2x1 - b4) / (x5 - x0)
        except:
            print()
    else:
        try:
            b0 = fx.subs(x, x0)
            b1 = (fx.subs(x, x1) - b0) / (x1 - x0)
            fx2x1 = ((fx.subs(x, x2) - fx.subs(x, x1)) / (x2 - x1))
            b2 = ((fx2x1 - b1) / (x2 - x0))
            fx3x2 = ((fx.subs(x, x3)) - fx.subs(x, x2)) / (x3 - x2)
            fx3x2x1 = (fx3x2 - fx2x1) / (x3 - x1)
            b3 = ((fx3x2x1 - b2) / (x3 - x0))
            fx4x3 = (fx.subs(x, x4) - fx.subs(x, x3)) / (x4 - x3)
            fx4x3x2 = (fx4x3 - fx3x2) / (x4 - x2)
            fx4x3x2x1 = (fx4x3x2 - fx3x2x1) / (x4 - x1)
            b4 = ((fx4x3x2x1 - b3) / (x4 - x0))
            fx5x4 = (fx.subs(x5) - fx.subs(x, x4)) / x5 - x4
            fx5x4x3 = (fx5x4 - fx4x3) / (x5 - x3)
            fx5x4x3x2 = (fx5x4x3 - fx4x3x2) / (x5 - x2)
            fx5x4x3x2x1 = (fx5x4x3x2 - fx4x3x2x1) / (x5 - x1)
            b5 = (fx5x4x3x2x1 - b4) / (x5 - x0)
        except:
            print()
    if grado == 1:
        polinomio = b0 + b1 * (x - x0)
        resultado.append([b0])
    elif grado == 2:
        polinomio = b0 + b1 * (x - x0) + (b2 * expand((x - x0) * (x - x1)))
        resultado.append([b0, b1, b2])
    elif grado == 3:
        polinomio = b0 + b1 * (x - x0) + (b2 * expand((x - x0) * (x - x1))) + (
                    b3 * expand((x - x0) * (x - x1) * (x - x2)))
        resultado.append([b0, b1, b2, b3])
    elif grado == 4:
        polinomio = b0 + b1 * (x - x0) + (b2 * expand((x - x0) * (x - x1))) + (
                    b3 * expand((x - x0) * (x - x1) * (x - x2))) + (
                                b4 * expand((x - x0) * (x - x1) * (x - x2) * (x - x3)))
        resultado.append([b0, b1, b2, b3, b4])
    elif grado == 5:
        polinomio = b0 + b1 * (x - x0) + (b2 * expand((x - x0) * (x - x1))) + (
                    b3 * expand((x - x0) * (x - x1) * (x - x2))) + (
                                b4 * expand((x - x0) * (x - x1) * (x - x2) * (x - x3))) + (
                                b5 * expand((x - x0) * (x - x1) * (x - x2) * (x - x3) * (x - x4)))
        resultado.append([b0, b1, b2, b3, b4, b5])
    interpolacion = polinomio.subs(x, interpo)
    if interpo != 0:
        resultado.append([polinomio, interpolacion])
    else:
        resultado.append([polinomio])

    return resultado


# La entrada funciona para 2 casos que te den una funcion o solo te den los valores de los "y"
# Si tienes la funcion no es necesario meter los valores de Y
# Tambien si no tienes todos los valores de "X" o "Y" ponle ceros
# La entrada va asi ("Funcion si hay si no hay pones 0",x0,x1,x2,x3,x4,x5,y0,y1,y2,y3,y4,y5,"El numero del grado del polinomio que quieras encontras")
lista = []
lista = polinomonewton(0, 4, -4, 7, 6, 2, 0, 278, -242, 1430, 908, 40, 0, 3, 5)
#la primera linea son las X, la segunda son las Y, la tercera son todoslos  B y la ultima el polinomio y la interpolacion
print(tabulate(lista))
```
